```python
import pandas as pd
import matplotlib.pyplot as plt

DATA_PATH = "../data/raw/ml-100k/"
```

```python
genres = pd.read_csv(DATA_PATH + "u.genre", sep="|", header=None, names=["genre", "id"])
genres_list = genres["genre"].to_list()

user_cols = ["user_id", "age", "sex", "occupation", "zip_code"]
users = pd.read_csv(DATA_PATH + "u.user", sep="|", names=user_cols)

item_cols = [
    "movie_id",
    "movie_title",
    "release_date",
    "video_date",
    "imdb_url",
] + genres_list
items = pd.read_csv(
    DATA_PATH + "u.item", sep="|", encoding="latin-1", header=None, names=item_cols
)
items = items.drop(["video_date", "imdb_url"], axis=1)

ratings_cols = ["user_id", "item_id", "rating", "timestamp"]
ratings = pd.read_csv(DATA_PATH + "u.data", sep="\t", names=ratings_cols).drop(
    columns="timestamp"
)
```

```python
ratings["rating"].hist(grid=False, bins=5, ec='black')
plt.xlabel("Rating")
plt.title("Ratings distribution")

# Customize x-axis labels
plt.xticks(range(1, 6))

plt.show()
```

```python
genres_count = dict.fromkeys(genres_list, 0)

for row in items.iterrows():
    row = row[1].to_dict()
    for item in row.items():
        key, value = item
        if key in genres_count.keys():
            genres_count[key] += value

plt.figure(figsize=(20, 10))
plt.bar(x=genres_count.keys(), height=genres_count.values())
plt.grid()
plt.show()
```

```python
# most popular films (with most of rankings)
popularity = ratings.groupby("item_id").count().sort_values("user_id", ascending=False)
popularity.reset_index(inplace=True)

mean_ratings = ratings.groupby("item_id").mean()['rating'].reset_index()

n_pop = 5
for i in range(n_pop):
    id = popularity.iloc[i]["item_id"]
    ranks = popularity["rating"]
    print(
        f"{i+1} Popular film is {items[items['movie_id'] == id].iloc[0]['movie_title']} "
        f"with {popularity[popularity['item_id'] == id].iloc[0]['rating']} rankings "
        f"with mean rating: {mean_ratings[mean_ratings['item_id'] == id].iloc[0]['rating']:.2f}"
    )
```

```python
popularity['rating'].plot(linewidth=3)
plt.ylabel('number of marks')
plt.title('Films popularity distribution')
plt.show()
```

```python
movie_ratings_dist = []
for movie in items.iterrows():
    r = mean_ratings[mean_ratings['item_id'] == movie[1]['movie_id']].iloc[0]['rating']
    movie_ratings_dist.append(r)

plt.hist(movie_ratings_dist, bins=100)
plt.grid()
plt.title('Movie ratings distribution')
plt.show()
```

```python
num_of_marks_users = ratings.groupby('user_id').count()['item_id']
```

```python
num_of_marks_users.hist(bins=100)
plt.title('Distribution of users ranks')
plt.show()
```

```python
print(f"minimum user ranks {num_of_marks_users.min()}")
print(f"max user ranks {num_of_marks_users.max()}")
```
